# Data Visualization using Seaborn

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
plt.style.use('classic')
x= np.linspace(-10,10,100)
plt.plot(x,np.sin(x));
```

```python
plt.style.use('classic')
x= np.linspace(-10,10,100)
sns.set()
plt.plot(x,np.sin(x));
```

```python
sns.get_dataset_names()
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
df.tail()
```

```python
sns.distplot(df['tip']);
```

```python
sns.distplot(df['tip'],kde=False,hist=True);
```

```python
sns.distplot(df['tip'],kde=True,hist=False);
```

```python
sns.jointplot(x='tip',y='total_bill',data=df);
```

```python
# corelation between each numerical variable with eachother
sns.pairplot(df);
```

```python
# corelation between each numerical variable with eachother
sns.pairplot(df,hue='sex');
```

```python
df.columns
```

```python
sns.pairplot(df,hue='smoker');
```

```python
# Plot among catergorical plot-establish relationship between  numeric and non numeric

sns.stripplot(x='time',y='tip',data=df);
```

```python
sns.boxplot(x='time',y='tip',data=df);
```

```python
sns.boxplot(x='sex',y='tip',data=df);
```

```python
sns.boxplot(x='smoker',y='tip',data=df);
```

```python
sns.barplot(x='smoker',y='tip',data=df);
```

```python
sns.barplot(x='sex',y='tip',data=df);
```
